```python
import shutil
import os
from glob import glob
from tqdm import tqdm

############# 저장 경로 수정 필요 #############
# img_source_dir = 'C:\\Users\\HP\\Desktop\\jupyter\\미니프로젝트2\\emotion_cv2\\final_data\\총합\\'
# save_dir = 'C:\\Users\\HP\\Desktop\\jupyter\\미니프로젝트2\\emotion_cv2\\final_data\\'
# image_list = glob('C:\\Users\\HP\\Desktop\\jupyter\\미니프로젝트2\\emotion_cv2\\final_data\\총합\\*.jpg')

img_source_dir = r'C:\Users\HP\Desktop\datasets\2020-02-092.한국인감정인식_sample\emotion_cv2\\'
# save_dir 만들기
save_dir = r'C:\Users\HP\Desktop\datasets\2020-02-092.한국인감정인식_sample\final_data\\'
# os.makedirs(save_dir)

image_list = glob(img_source_dir +'*.jpg')

############# 저장 경로 수정 필요 #############


class_name = ['기쁨','당황','분노','불안','상처','슬픔','중립']
folder_names = ['기쁨','당황','분노','불안','상처','슬픔','중립','재분류필요']

for foldername in folder_names:
    os.mkdir(save_dir + foldername)
for img in tqdm(image_list):
    # 이미지 있는지 체크
    
    if os.path.isfile(img):
        # max값 찾아서 index에 저장하고 몇번 max인지 횟수도 저장
        
        file_name = img[68:]
        df_file_name = str(file_name[:65] + file_name[-11:])
        row = final_df_all[final_df_all.index == df_file_name]
        row_list = list(row.values[0])

        max_index = -1
        max_index_count = 0
        for cnt2, i in enumerate(row_list):
            max_vote = vote_list[vote_list.index== df_file_name].values[0]
            if max_vote == i:
                max_index = cnt2
                max_index_count += 1

        ## 최종적으로 이미지 복사
        if max_index_count == 1:
            img_folder_name = class_name[max_index]
            shutil.copy(img, save_dir + img_folder_name)
        else :
            shutil.copy(img, save_dir + "재분류필요" )
```
